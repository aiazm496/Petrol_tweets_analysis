```python
import tweepy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import credentials
```

```python
auth = tweepy.OAuthHandler(credentials.CONSUMER_KEY,credentials.CONSUMER_SECRET)
```

```python
auth.set_access_token(credentials.ACCESS_TOKEN,credentials.ACCESS_TOKEN_SECRET)
```

```python
api = tweepy.API(auth)
```

```python
tweet_data = api.search('petrol',count = 500)
```

```python
tweet_data
```

```python
tweet_data[0]
```

```python
dir(tweet_data[0])
```

```python
user = []
text = []
retweets = []
favourites = []



for tweet in tweet_data:
    user.append(tweet.user.name)
    text.append(tweet.text)
    retweets.append(tweet.retweet_count)
    favourites.append(tweet.favorite_count)
    
    
    
```

```python
pet_dis_df = pd.DataFrame({'user':user,'text':text,'retweets':retweets,'favourites':favourites})
```

```python
pet_dis_df.head()
```

```python
pet_dis_df_analysis = pet_dis_df.copy(deep=True) #orginal df safe from slicing changes
```

```python
pet_dis_df_analysis.sort_values(by = 'retweets',inplace=True,ascending=False)
```

```python
pet_dis_df_analysis.head()
```

```python
for texts in pet_dis_df_analysis.text:
    print(texts)
    print('*'*100)
```

```python
#semantic analysis of 100 tweets on petrol
```

```python
from textblob import TextBlob
```

```python
neg = 0
pos = 0
neu = 0

for tweet in tweet_data:
    blob = TextBlob(tweet.text)
    if blob.sentiment.polarity < 0:
        neg+=1
        
    elif blob.sentiment.polarity == 0:
        neu+=1
        
    else:
        pos+=1
        
```

```python
plt.pie(x = [neg,pos,neu],labels=['negative','positive','neutral'],autopct='%1.1f%%')
```

```python
print("The percentage of negative tweets in 50 tweets: ", neg*100/50,'%')
```
